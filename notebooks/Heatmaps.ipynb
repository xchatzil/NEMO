```python
df_mean = pd.DataFrame.from_dict(data_mean, orient='index')
df_p90 = pd.DataFrame.from_dict(data_p90, orient='index')
```

```python
#Convert the DataFrame to integers
df_mean = df_mean.astype(int)
df_p90 = df_p90.astype(int)
```

```python
# create the heatmap
df = df_mean.copy()
plt.figure(figsize=(14, 6))
sns.set(font_scale=1.2)
sns.heatmap(df, annot=True, cmap='viridis_r', vmax=500, cbar_kws={'label': 'latency (ms)'}, xticklabels=columns,
            cbar=True, yticklabels=rows, fmt='g')

# Rotate xtick labels
plt.xticks(rotation=35, ha='right')

# show the plot
plt.savefig("plots/heatmap_means.svg", bbox_inches='tight')
plt.title("Average Latency")
plt.show()
```

```python
# create the heatmap
df = df_p90.copy()
plt.figure(figsize=(14, 6))
sns.set(font_scale=1)
sns.heatmap(df, annot=True, cmap='viridis_r', vmax=500, cbar_kws={'label': 'latency (ms)'}, xticklabels=columns,
            cbar=True, yticklabels=rows, fmt='g')

# Rotate xtick labels
plt.xticks(rotation=22, ha='right')

# show the plot
plt.savefig("plots/heatmap_p90.svg", bbox_inches='tight')
plt.title("P90 Latency")
plt.show()
```

```python
# create the heatmap
df = df_mean.copy()
df = df.div(df.min(axis=1), axis=0)
df = df.drop("dt", axis=1)
new_cols = columns[1:]

plt.figure(figsize=(14, 6))
sns.set(font_scale=1.2)
sns.heatmap(df, annot=True, cmap='Greens', vmax=3, cbar_kws={'label': 'latency (ms)'}, cbar=True, xticklabels=new_cols,
            yticklabels=rows, fmt='.2f')

# Rotate xtick labels
plt.xticks(rotation=35, ha='right')

# show the plot
plt.savefig("plots/heatmap_means_norm.svg", bbox_inches='tight')
plt.title("Normalized Average Latency")
plt.show()
```
