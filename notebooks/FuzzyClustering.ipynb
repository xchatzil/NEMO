```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt
```

```python
# Generate synthetic data
np.random.seed(42)
data = np.vstack([np.random.normal(0, 1, (100, 2)),
                  np.random.normal(4, 1, (100, 2)),
                  np.random.normal(8, 1, (100, 2))])

# Apply Fuzzy C-Means clustering
num_clusters = 2
centers, u, u0, d, jm, p, fpc = fuzz.cluster.cmeans(data.T, c=num_clusters, m=2, error=0.005, maxiter=1000, init=None)

# Membership matrix (u) contains the membership degrees for each data point
# Cluster centers (centers) represent the final cluster centers

# Plot the results
cluster_membership = np.argmax(u, axis=0)
colors = ['b.', 'g.', 'r.']
for j in range(3):
    plt.plot(data[cluster_membership == j, 0],
             data[cluster_membership == j, 1], colors[j])

for pt in centers.T:
    plt.plot(pt[0], pt[1], 'yo', markersize=8)

plt.title('Fuzzy C-Means Clustering')
plt.show()
```
